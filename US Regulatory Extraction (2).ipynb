```python
from bs4 import BeautifulSoup
import requests,json
import pandas as pd
```

```python
# Through Request received the API Key to extract data from https://open.gsa.gov/api/regulationsgov/
# Using requests package we extract the html content we further convert the data as readable using beautiful soup package
api_key = "qGZagngC6YxQ0mjBWRaZkS0xxLtP128IYII6C8Mu"
url = "https://api.regulations.gov/v4/dockets?filter[searchTerm]=Tobacco Products&api_key=" + api_key # creating the url
result=requests.get(url,headers={"User-Agent":"Mozilla/5.0"})
url_content = BeautifulSoup(result.text,'lxml')# Request for data
```

```python
Json_data = json.loads(url_content.html.body.text)
Json_data['data']
```

```python
# Extracting the doket id and object ids 
docket_id = [i['id'] for i in Json_data['data']]
docket_object_id = [i['attributes']['objectId'] for i in Json_data['data']]
```

```python
docket_object_id
```

```python
# This cell will go through all the dockets and get their documents ids and object ids
document_ids = []
document_objectId = []
for dockt in docket_id:
    url = "https://api.regulations.gov/v4/documents?filter[docketId]="+dockt+"&api_key=" + api_key
    url_content = BeautifulSoup(requests.get(url).content, 'lxml')
    data = json.loads(url_content.body.p.text)["data"]
    document_ids.extend([i['id'] for i in data]) # get document ids
    document_objectId.extend([i['attributes']['objectId'] for i in data]) # get document object ids 
```

```python
list(set(document_ids))
```

```python
list(set(document_objectId))
```

```python
document_ids = list(set(document_ids)) # removing similar ids 
document_objectId = list(set(document_objectId)) # Removing similar object ids of documents
```

```python
# This cell will get all the comments ids from the above scrapped document object ids
commentsIds = []
for doc_objectId in document_objectId:
    for num in range(1, 35): # check for 35 pages for each doc id
        url = "https://api.regulations.gov/v4/comments?filter[commentOnId]="+doc_objectId+"&page[size]=250&page[number]="+str(num)+"&sort=lastModifiedDate,documentId&api_key="+ api_key
        url_content = BeautifulSoup(requests.get(url).content, 'lxml')
        print(url_content)
        try:
            comment_json = json.loads(url_content.body.p.text)['data'] 
            if comment_json==[]: break # break the loop if their is seq of empty pages
            commentsIds.append([i['id'] for i in comment_json]) 
        except: break
commentsIds = list(set([j for i in commentsIds for j in i if j!=[]])) # remove similar comment ids
```

```python
# this cell will get comments from each comment id and store it into dict formate
comments = {} #declaring dictionary
for comment_id in commentsIds:
    url = "https://api.regulations.gov/v4/comments/"+comment_id+"?include=attachments&api_key="+ api_key
    url_content = BeautifulSoup(requests.get(url).content, 'lxml')
    try: comments[comment_id]= json.loads(url_content.body.p.text)["data"]["attributes"]["comment"]
    except: pass
```

```python
# Save the comments
df = pd.DataFrame(list(comments.items()), columns=["commentIds", 'comments'])
df
```
